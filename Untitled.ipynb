```python
def normalize(array):
    out_array = array.copy()
    for n in xrange(array.shape[0]):
        out_array[n] -= out_array[n].mean()
        out_array[n] /= array[n].std()
    return out_array
```

```python
def normalize(array):
    out = np.zeros(array.shape)
    for n in xrange(array.shape[0]):
        out[n] = (array[n]-array[n].mean())/array[n].std()
    return out

def expand(i,stride):
    out = np.zeros((stride,i.shape[0]))
    for n in xrange(stride):
        out[n] = i
    return out

def xz_stack(stack,images=2,stride=6):
    xz_stack = stack[:,:images,:]

    xz_avg = np.zeros((xz_stack.shape[1],xz_stack.shape[0]*stride,xz_stack.shape[2]))

    for n in xrange(xz_avg.shape[0]):
        for m in xrange(xz_stack.shape[0]):
            xz_avg[n,stride*m:stride*(m+1)] = expand(xz_stack[m,n],stride)

    return xz_avg
```

```python
img = xz_stack(img_norm)[0]
img_alt = img.copy()

for n in xrange(img_alt.shape[0]):
    img_alt[n] = (img_alt[n] - img_alt[n].mean())/img_alt[n].std()
```

```python
plt.figure(figsize=(9,9))
plt.imshow(img,cmap=cmap)
plt.figure(figsize=(9,9))
plt.imshow(img_alt,cmap=cmap)
```
